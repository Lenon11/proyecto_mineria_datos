```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv("../data/raw/IIEG_Seguridad.csv") 
```

```python


# ===============================
# 1. Selección de columnas útiles
# ===============================
cols_keep = [
    "fecha",
    "delito",
    "municipio",
    "clave_mun",
    "bien_afectado",
    "zona_geografica"
]

df = df[cols_keep].copy()

# ===============================
# 2. Limpieza de municipio
# ===============================
df = df[
    (df["municipio"] != "N.D.") &
    (df["municipio"].notna()) &
    (df["clave_mun"] != "999")
]

# ===============================
# 3. Conversión de fecha
# ===============================
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")

# ===============================
# 4. Features temporales
# ===============================
df["year"] = df["fecha"].dt.year
df["month"] = df["fecha"].dt.month

# ===============================
# 5. Normalizar Strings
# ===============================
df["municipio"] = df["municipio"].str.strip().str.upper()
df["delito"] = df["delito"].str.strip().str.upper()
# ===============================
# 6. Limpieza final
# ===============================
df = df.dropna(subset=["fecha", "delito", "municipio"])

df.reset_index(drop=True, inplace=True)
```
